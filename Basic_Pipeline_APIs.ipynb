Sentiment Analysis

```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis')
```

```python
classifier("That song was psychedelic")
```

# Zero Shot classification

```python
zer = pipeline("zero-shot-classification")
```

```python
zer("Was this beat about france?", candidate_labels = ["music", "biology", "art"])
```

# Text Generation

```python
gen = pipeline("text-generation", model = "distilgpt2")
```

```python
gen("In a fight between Gandhi and MLK, the winner would be", max_length = 30, num_return_sequences = 2)
```

# Filling Blanks

```python
fill_blank = pipeline("fill-mask")
```

```python
fill_blank("With great <mask> comes great responsibilty", top_k=2)
```

# Summarization

```python
summarizer = pipeline("summarization")
```

```python
summarizer("It is a period of civil war. Rebel spaceships, striking from a hidden base, have won their first victory against the evil Galactic Empire. During the battle, Rebel spies managed to steal secret plans to the Empire’s ultimate weapon, the DEATH STAR, an armored space station with enough power to destroy an entire planet. Pursued by the Empire’s sinister agents, Princess Leia races home aboard her starship, custodian of the stolen plans that can save her people and restore freedom to the galaxy….",
           min_length = 5,
           max_length = 20)
```

# Named Entity Recognition

```python
ner = pipeline("ner", grouped_entities=True)
```

```python
ner("Cal Lynch travels back in time to 15th-century Spain to defeat the Templars")
```
